# CP functions

```python
import numpy as np
import matplotlib.pyplot as plt
#from matplotlib import cm
%matplotlib notebook
```

* Data and function for plotting

```python
def plotcontour(f, n=201, a=2):
    k = np.linspace(-a, a, n)
    x, y = np.meshgrid(k, k)
    z = f(x, y)
    
    fig = plt.figure()
    CS = plt.contour(x, y, z)
    plt.clabel(CS, inline=1, fontsize=10)
```

## Fischer function
$$\phi_F(u, v) = \sqrt{u^2 + v^2} -u - v$$

```python
def Fischer(u, v): 
    return np.sqrt(u**2 +v**2) - u - v

plotcontour(Fischer)
```

## Minimum
$$\phi_{NR}(u,v) = \min(u, v)$$

```python
plotcontour(np.minimum)
```

## Mangasarian and Solodov function
$$\phi_{Reg}(u,v) = uv + \frac{1}{2\alpha}[\max(0, u-\alpha v)^2 - u^2] $$

```python
def msf(u, v, alpha=1):
    return u*v + (np.maximum(0, u - alpha*v)**2 - u**2)/(2*alpha)

plotcontour(msf)
```
